```python
##################################################################
 # Nama : M. Mahfi Nurandi Karsana                              #
 # NIM  : 1301174473                                            #
 # Kelas: IF-41-10                                              #
 # MOSI 4 : Random Walk                                         # 
##################################################################
 #                                                              #
 # Note : only work for windows                                 #
 #                                                              #
##################################################################
# import
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt


# list that carry object
obj_list = []

# object number
n_object = 10

# init object
for i in range(n_object) :
    obj_list.append([])
```

```python
# init random starting point
for i in range(n_object) :
    # temp container for starting cordinate
    cor = []
    # 2 dimension
    for j in range(0,2) :
        # starting point x = 0 - 10
        # y = 0 - 10
        cor.append(np.random.randint(0,10, dtype='int'))
    # add to list of object
    obj_list[i].append(cor)
```

```python
# num of iter
num_iter = 100

# main loop iteration 
for i in range(0,num_iter) :
    
    # iterating through object in obj_list
    for obj_idx in range(0,10) :
        # starting point
        starting_point = obj_list[obj_idx][-1]
        
        # determine moves up down left right
        # random float
        random_number = np.random.rand()
        
        # copy starting point to another variable
        endPoint = starting_point[:]
        
        # up
        if 0 <= random_number and random_number <= 0.25 :
            endPoint[1] = endPoint[1] +1
        # down
        elif 0.26 <= random_number and random_number <= 0.50 :
            endPoint[1] = endPoint[1] -1
        # left
        elif 0.51 <= random_number and random_number <= 0.75 :
            endPoint[0] = endPoint[0] -1
        # right
        elif 0.75 <= random_number and random_number <= 1 :
            endPoint[0] = endPoint[0] +1
        
        if endPoint[0] < 0 :
            endPoint[0] = endPoint[0] +1
        elif endPoint[0] > 20 :
            endPoint[0] = endPoint[0] -1
        
        if endPoint[1] < 0 :
            endPoint[1] = endPoint[1] +1
        elif endPoint[1] > 20 :
            endPoint[1] = endPoint[1] -1
        
        
        # assign endpoint
        # append to end of object
        obj_list[obj_idx].append(endPoint)

        
```

```python
# get min max x , y
x_axis = []
y_axis = []

# iterating through object in obj_list
for i in range(10) :
    # iterating through every element in the obj_list
    for elm in obj_list[i] :
        # add x,  and y to a temporary container
        x_axis.append(elm[0])
        y_axis.append(elm[1])

# get min max of x and y 
# -1 and +1 to make it look better
lim_x = [min(x_axis)-1, max(x_axis)+1]
lim_y = [min(y_axis)-1, max(y_axis)+1]

lim_x = [-1, 21]
lim_y = [-1, 21]
```

```python

# colour for every object
colour_code = ['red', 'green', 'blue', 'brown', 'orange', 'purple', 'black', 'gold', 'crimson', 'limegreen']
# iterating for the length of iter + 1 
# +1 because starting point is different than iter
for j in range(len(obj_list[0])) :
    # make a matplot figure
    fig , ax = plt.subplots()
    # add title to the figure
    fig.suptitle('iter %d' % j)
    
    # iterating through the object
    # so that every object will print
    # the same idx of elm in obj
    for i in range(10) :
        # add a point to figure
        ax.scatter(x=obj_list[i][j][0], y=obj_list[i][j][1], color=colour_code[i])
        # annotate every point with obj index in obj_list 
        ax.annotate(str(i),(obj_list[i][j][0],obj_list[i][j][1]))
    
    # add grid
    ax.grid(True)
    
    # add axis laber 
    plt.xlabel('x')
    plt.ylabel('y')
    
    # limit axis with teh value previously gotten
    plt.xlim(lim_x)
    plt.ylim(lim_y)
    
    # save to a file
    plt.savefig('iter_%d.png' % j)
    
    # closing the figure
    plt.close(fig)

```
